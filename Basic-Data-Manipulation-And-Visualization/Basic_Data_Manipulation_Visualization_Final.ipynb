### 80. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries
Hint: First we need to find top-3 countries based on revenue, then create scater plot between Quantity and Unitprice for each country separately


```python
# there is an outlier with negative price
online_rt = online_rt[online_rt.UnitPrice>0]
```

```python
# Calculating the Revenue
online_rt['Revenue'] = online_rt['Quantity']*online_rt['UnitPrice']
```

```python
#top three countries based on revenue
top3 = online_rt.groupby('Country').Revenue.sum().sort_values(ascending = False).index[:3]
```

```python
top3
```

```python
#A single customer can have any number of orders with different prices, so summing up prices isn't quite helpful.
#A reasonable alternative is that we assign each customer the average amount each has paid per item.
```

```python
grouping_customers = online_rt[online_rt.Country.isin(top3)].groupby(['CustomerID', 'Country'] )
plottable = grouping_customers['Quantity','Revenue'].agg('sum')
plottable['AvgPrice'] = plottable.Revenue / plottable.Quantity

# get the value of the index and put in the column Country
plottable['Country'] = plottable.index.get_level_values(1)
plottable.head()
```

```python
# creates the FaceGrid
g = sns.FacetGrid(plottable, col="Country")

# map over a make a scatterplot
g.map(plt.scatter, "Quantity", "AvgPrice", alpha=1)

# adds legend
g.add_legend();
```
